```python
from bs4 import BeautifulSoup
import requests
import re


dic2=["D000","A300","A200","A100","B000","C100","C200","C300","C400","L000","E000","J000","I000","K000"
    ,"F000","G000","H000","1000","M000","N000","O000","2000","Q000","3000","R000","S000","T000","U000"
    ,"V000","W000","Y000","X000"]
dic={u"上游",u"中游",u"下游"}
f = open("E:\\python\\chhhhhain.txt","w")

for i in dic2:
    lis = []
    c_count = 0 #計算company
    t_count = 0 #計算title
    s_count = 0 #計算subchain
    b_count = 0 #計算公司companylist
    #to_count = 0 #計算totla

    page_url = "http://ic.gretai.org.tw/introduce.php?ic={}"
    res = requests.get(page_url.format(i))
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text)
    entries  = soup.select("div")
    r = re.findall("companyList_.+?\"",str(entries))

    com_title  = soup.select(".chain-title-panel") #上中下游
    com_chain  = soup.select(".chain") #產業類別
    com_total  = soup.select(".company-number") #共?家
    header = soup.select(".page-header")
    content = ''
    text =''


    #print type(r)
    text += header[0].text+"\n"+"=================="+"\n"

    for i in range(0,len(r)/6):
         lis.append(r[i])

    #len(lis)
    if com_title !=[]:
        for i in com_chain: 
            content += i.text

        for title in range(0,len(content.split())): #上中下游 len(content.split())
            if content.split()[title] not in dic: #顯示class
                text  +=  content.split()[title]+","
                #for i in range(0,2):
                c_count = 0
                s = "#"+lis[title-1].split("\"")[0]
                companylist = soup.select(s+" .company b")    
                company = soup.select(s+" .company a")
                com_total  = soup.select(".company-number") #共?家
                sub = soup.select("#sc_industry_"+s.split("_")[1]+" div")   
                com_total  = soup.select(s+" .company-number") #共?家
                
                if com_total !=[]:
                    text += com_total[0].text+"\n"

                if sub !=[]: #判斷有無subchain
                    for i in range(0,len(sub)): #len(sub)
                        text += "".join(sub[i].text.split()) +"\n"
                        s_total = int(re.findall("[0-9]+","".join(sub[i].text.split()))[0])

                        #print "".join(sub[i].text.split())

                        while s_count < s_total:   #市場別
                            text += companylist[b_count].text +","
                            total = int(re.findall("[0-9]+",companylist[b_count].text)[0])

                            #print companylist[b_count].text

                                #print total
                            for i in range(0,total): #公司名
                                text += company[c_count].text
                                if i < total-1:
                                    text += ","
                                else:
                                    text += "\n"
                                c_count +=1
                                s_count +=1
                            b_count +=1
                        s_count = 0   
                        text += "-----------------\n"
                    b_count=0

                else:    
                    for i in range(0,len(companylist)/2): #市場別
                        text += companylist[i].text +","
                        total = int(re.findall("[0-9]+",companylist[i].text)[0])

                        #print total
                        for i in range(0,total): #公司名
                            text += company[c_count].text
                            if i < total-1:
                                text += ","
                            else:
                                text += "\n"
                            c_count +=1
                    text += "-----------------\n"
            else:#顯示上下游
                text  +=  content.split()[title]+"\n"
                text += "~~~~~~~~~~~~~~~~~~~~~~~~~~~\n"
                        #text += "\n" #只是換個行
                
    else:
        com_chain  = soup.select(".chain-company")       
        for i in range(0,len(com_chain)): #市場別
            s = "#"+lis[i].split("\"")[0]
            c_count = 0
            companylist = soup.select(s+" .company b")    
            company = soup.select(s+" .company a")
            sub = soup.select("#sc_industry_"+s.split("_")[1]+" div")   
            com_total  = soup.select(s+" .company-number") #共?家
            text += "".join(com_chain[i].text.split()) +","
            
            if com_total !=[]:
                text += com_total[0].text+"\n"
            
            for i in range(0,len(companylist)/2): #市場別
                text += companylist[i].text +","
                total = int(re.findall("[0-9]+",companylist[i].text)[0])
                #print total
                for i in range(0,total): #公司名
                    text += company[c_count].text
                    if i < total-1:
                         text += ","
                    else:
                        text += "\n"
                    c_count +=1
                text += "-----------------\n" 
    #最大迴圈結束
    f.write(text.encode("utf-8"))

f.close()
```

```python
from bs4 import BeautifulSoup
import requests,re,json



dic={"CHAIRMAN_NAME","COMPANY_ABBR_NAME","COMPANY_ADDRESS","COMPANY_TEL","INTERNET_ADDRESS"
    ,"NAME","PRESIDENT_NAME","SPOKENMAN","MAIN_BUSINESS1","MAIN_BUSINESS2","MAIN_BUSINESS3"
    ,"LISTING_DATE"}
dic2=["D000","A300","A200","A100","B000","C100","C200","C300","C400","L000","E000","J000","I000","K000"
    ,"F000","G000","H000","1000","M000","N000","O000","2000","Q000","3000","R000","S000","T000","U000"
    ,"V000","W000","Y000","X000"]
dic3=[]
basic_url = "http://ic.gretai.org.tw/company_basic.php?stk_code={}"
f = open("E:\\python\\chain\\basic_info.txt","w")




for j in dic2:
    page_url = "http://ic.gretai.org.tw/introduce.php?ic={}"
    res = requests.get(page_url.format(j))
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text)
    link  = soup.select(".company a")
    js_content =''
    content=''
    content2=''
    
    for i in range(0,len(link)/2):
        #print link[i]['href'].split("=")[1]
        a = link[i]['href'].split("=")[1]
        
        if a not in dic3:
            dic3.append(a)
            res = requests.get(basic_url.format(a))
            res.encoding = 'utf-8'
            soup = BeautifulSoup(res.text)
            item = soup.select("tbody td")
            header  = soup.select(".page-header")

            data = soup.select("script")
            for i in data:
                js_content += i.text

            r = re.findall("http://www.gretai.org.tw/storage/company_basic/.+",js_content)[0].split("\"")[0]
            js_content =''
            print a
            print r

            js = requests.get(r).text

            if js =="":
                content += header[0].text+",Null\n"
            else:
                orgTest = js.encode("utf-8").split("getCompanyBasic(")[1].split("})")[0]+"}"
                json1 = json.loads(orgTest)
                content += header[0].text+","

                for data in json1:
                    if data in dic:
                        content += json1[data]+","

                for i in item:
                    content2 += i.text
                content += "".join(content2.split())+"\n"

            f.write(content.encode("utf-8"))
            content =''
            content2 =''
    
f.close()
    
```
