```python
from bs4 import BeautifulSoup
import requests
import re

page_url = "http://cn.wsj.com/big5/{}.asp"
dic=["bus","bas","beu","japan","nkorea","russia","baf","latina"]
country=["美國","亞太","歐洲","日本","朝鮮","俄羅斯","非洲","拉丁美洲"]

#print(len(item.select('a')[5]))
#print(item.select('a')[5])


for page in range(0,len(dic)):
    res = requests.get(page_url.format(dic[page])) 
    res.encoding = 'big5'
    soup = BeautifulSoup(res.text)
    entries  = soup.select("#top2left")
    print("-------------------"+country[page]+"---------------------")
    
    for item in entries:
        for i in range(0,len(item.select('a'))):
            m = re.search('.+',item.select('a')[i].text)
            if m:
                href = item.select('a')[i]['href']
                print(item.select('a')[i].text)
                print ('http://cn.wsj.com/big5/' + href)

```
